Q1

```python
import requests
import pickle
from bs4 import BeautifulSoup

url = "https://web.archive.org/web/20210327165005/https://hackaday.com/2021/03/22/how-laser-headlights-work/"

# Extract the HTML from the URL
response = requests.get(url)
html_content = response.text

# Save the HTML to a .pkl file
with open("article_html.pkl", "wb") as file:
    pickle.dump(html_content, file)
```

Q2

```python
with open("article_html.pkl", "rb") as file:
    html_content = pickle.load(file)
    soup = BeautifulSoup(html_content, "html.parser")
    article_text = soup.get_text()
    print(article_text)
```

Q3

```python
import spacy
from collections import Counter

# Load the spaCy pipeline (make sure you have downloaded the language model)
nlp = spacy.load("en_core_web_sm")

# Process the article text using spaCy
doc = nlp(article_text)

# Filter out punctuation, stopwords, and whitespace, and convert tokens to lowercase
tokens = [token.text.lower() for token in doc if not token.is_punct and not token.is_stop and not token.is_space]

# Find the 5 most frequent tokens
most_common_tokens = Counter(tokens).most_common(5)

# Print the most common tokens and their frequencies
print("Most Common Tokens:")
for token, frequency in most_common_tokens:
    print(f"{token}: {frequency}")
```

Q4

```python
# Filter out punctuation, stopwords, and whitespace, and convert lemmas to lowercase
lemmas = [token.lemma_.lower() for token in doc if not token.is_punct and not token.is_stop and not token.is_space]

# Find the 5 most frequent lemmas
most_common_lemmas = Counter(lemmas).most_common(5)

# Print the most common lemmas and their frequencies
print("Most Common Lemmas:")
for lemma, frequency in most_common_lemmas:
    print(f"{lemma}: {frequency}")
```

Q5

```python
def score_sentence_by_token(sentence, interesting_tokens):
    words = sentence.split()
    num_words = len(words)
    interesting_count = sum(1 for word in words if word.lower() in interesting_tokens)
    return interesting_count / num_words

def score_sentence_by_lemma(sentence, interesting_lemmas):
    doc = nlp(sentence)
    lemmas = [token.lemma_.lower() for token in doc if not token.is_punct and not token.is_stop and not token.is_space]
    num_words = len(lemmas)
    interesting_count = sum(1 for lemma in lemmas if lemma in interesting_lemmas)
    return interesting_count / num_words

# Test the methods with the first sentence in the article and the most common tokens and lemmas from Question 3.
first_sentence = article_text.split(".")[0]  # Take the first sentence
most_common_tokens = [token for token, _ in most_common_tokens]
most_common_lemmas = [lemma for lemma, _ in most_common_lemmas]

token_score = score_sentence_by_token(first_sentence, most_common_tokens)
lemma_score = score_sentence_by_lemma(first_sentence, most_common_lemmas)

print(f"Token Score for the first sentence: {token_score}")
print(f"Lemma Score for the first sentence: {lemma_score}")
```

Q6

```python
import matplotlib.pyplot as plt

# Split the article text into sentences
sentences = [sent.strip() for sent in article_text.split(".") if sent.strip()]

# Calculate the scores for each sentence using the most common tokens from Question 3
scores_by_tokens = [score_sentence_by_token(sentence, most_common_tokens) for sentence in sentences]

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(scores_by_tokens, bins=20, edgecolor="black")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.title("Distribution of Scores for Sentences")
plt.show()
```
